```python
df1['num_tickets'] = df1.Ticket.apply(lambda x: 1 if x.isnumeric() else 0)
df1.num_tickets.value_counts()
```

```python
df1['cat_tickets'] = df1.Ticket.apply(lambda x: ''.join(x.split(' ')[:-1]).replace('.','').replace('/','').lower() if len(x.split(' ')[:-1])>0 else 0)
df1.cat_tickets.value_counts()
```

```python
df1['name_title'] = df1.Name.apply(lambda x: x.split(',')[1].split('.')[0].strip())
df1.name_title.value_counts()
```

```python
pd.pivot_table(df1, index='Survived', columns='name_title' , values='Ticket', aggfunc='count')
```

```python
df1.dropna(subset=['Embarked'], inplace=True)
```

```python
df1.Age = df1.Age.fillna(df1.Age.median())
```

```python
df1.drop(['Cabin'], axis=1, inplace=True)
```

```python
df1.drop(['Name'], axis=1, inplace=True)
```

```python
df1.drop(['PassengerId'], axis=1, inplace=True)
```

```python
df1.drop(['Ticket'], axis=1, inplace=True)
```

```python
df1.head()
```

```python
df1['norm_sibsp'] = np.log(df1.SibSp+1)
```

```python
df1['norm_sibsp'].hist()
```

```python
df1['norm_fare'] = np.log(df1.Fare+1)
df1['norm_fare'].hist()
```

```python
df1.Pclass = df1.Pclass.astype(str)
```

```python
df1.dtypes
```

```python
df1.drop(['SibSp', 'Fare'], axis=1, inplace=True)
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
dum_sex = pd.get_dummies(df1['Sex'], drop_first=True)
dum_sex
```

```python
dum_pclass = pd.get_dummies(df1['Pclass'], drop_first=True)
dum_pclass
```

```python
dum_emb = pd.get_dummies(df1['Embarked'], drop_first=True)
dum_emb
```

```python
dum_cab_letter = pd.get_dummies(df1['cabin_letter'], drop_first=True)
dum_cab_letter
```

```python
dum_cat_tickets = pd.get_dummies(df1['cat_tickets'], drop_first=True)
dum_cat_tickets
```

```python
dum_name_title = pd.get_dummies(df1['name_title'], drop_first=True)
dum_name_title
```

```python
df2 = pd.concat([df1, dum_sex, dum_pclass, dum_emb, dum_cab_letter, dum_cat_tickets, dum_name_title], axis=1)
df2.head()
```

```python
df2.drop(['Sex', 'Pclass', 'Embarked', 'cabin_letter', 'cat_tickets', 'name_title'], axis=1, inplace=True)
```

```python
df2.head()
```

```python
df3 = df2.copy()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
sc = StandardScaler()
mms = MinMaxScaler()
```

```python
df3[['Age','norm_sibsp','Parch','norm_fare']]= mms.fit_transform(df3[['Age','norm_sibsp','Parch','norm_fare']])
```

```python
df3.head()
```

```python
X = df3.iloc[:, 1:].values
y = df3.iloc[:, 0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train, y_train)
print(svc_model.score(X_test, y_test))
print(svc_model.score(X_train, y_train))
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
print(lr_model.score(X_test, y_test))
print(lr_model.score(X_train, y_train))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
print(dt_model.score(X_test, y_test))
print(dt_model.score(X_train, y_train))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
print(rf_model.score(X_test, y_test))
print(rf_model.score(X_train, y_train))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train, y_train)
print(gb_model.score(X_test, y_test))
print(gb_model.score(X_train, y_train))
```

```python
import xgboost
xgb_model = xgboost.XGBClassifier()
xgb_model.fit(X_train, y_train)
print(xgb_model.score(X_test, y_test))
print(xgb_model.score(X_train, y_train))
```

```python
lr_params = {
    'penalty':['l1','l2'],
    'C':np.logspace(-4,4,5)
}
```

```python
svc_params = {
    'C':[1,5,10,50,100],
    'kernel':['linear', 'poly', 'rbf'],
    'degree':[1,2,3],
    'gamma':['scale','auto']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
lr_model2 = LogisticRegression()
clf_lr = GridSearchCV(lr_model2, lr_params, n_jobs=1, verbose=2, cv=5)
clf_lr.fit(X,y)
```

```python
clf_lr.best_estimator_
```

```python
svc_model2 = SVC()
clf_svc = GridSearchCV(svc_model2, svc_params, n_jobs=1, verbose=2, cv=5)
clf_svc.fit(X,y)
```

```python
clf_svc.best_estimator_
```

```python
clf = SVC(C=10, kernel='poly')
clf.fit(X_train, y_train)
```

```python
print(clf.score(X_test, y_test))
print(clf.score(X_train, y_train))
```

```python
from sklearn.ensemble import ExtraTreesClassifier
mod = ExtraTreesClassifier()
mod.fit(X_train, y_train)
```

```python
imp_frame = pd.Series(mod.feature_importances_)
imp_frame.nlargest(15)
```

```python
from sklearn.feature_selection import SelectKBest, chi2
best = SelectKBest(score_func=chi2, k=15)
mod2 = best.fit(X, y)
```

```python
featscores = pd.DataFrame(mod2.scores_)
featnames = pd.DataFrame(df3.drop('Survived', axis=1).columns)
```

```python
df = pd.concat([featnames, featscores], axis=1)
df.columns = ['Feats', 'Score']
df.nlargest(20, 'Score')
```
